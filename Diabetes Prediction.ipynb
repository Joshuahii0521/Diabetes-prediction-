### 7.2.3 Remove outliers

```python
def iqr_outliers(df):
    out=[]
    first_quartile = df.quantile(0.25)
    third_quartile = df.quantile(0.75)
    inter_quartile_range = third_quartile-first_quartile
    Lower_tail = first_quartile - 1.5 * inter_quartile_range
    Upper_tail = third_quartile + 1.5 * inter_quartile_range
    for i in df:
        if i > Upper_tail or i < Lower_tail:
            out.append(i)
    print("Outliers:",len(out))
for col in df.drop('Outcome',axis=1).columns:
    iqr_outliers(df[col])
```

```python
for col in df.drop('Outcome',axis=1).columns:
    first_quartile = df[col].quantile(0.25)
    third_quartile = df[col].quantile(0.75)
    inter_quartile_range = third_quartile-first_quartile
    Lower_tail = first_quartile - 1.5 * inter_quartile_range
    Upper_tail = third_quartile + 1.5 * inter_quartile_range
    
    df[col] = np.where((df[col]<Lower_tail) | (df[col]>Upper_tail), df[col].median(),df[col])
```

We will replace the outlier with median, as removing them will cause loss of important informations.

# 11. Building the Model

## 11.1 Classification


### 11.1.1 Model and evaluation metrics selection

A total of four models will be used in the prediction process: Decision Tree, KNN, Random Forest, and XGBoost. The reason is that KNN and Decision Tree are two foundational but accurate algorithms for predicting diabetes. In order to see whether these algorithms can perform well, we simply use the two algorithms. We then explore Random Forest and XGBoost, which are two little advanced algorithms to see if performance can be further enhanced.

Since our dataset is unbalanced, accuracy is not suitable. The F1-score is a harmonic mean of Precision and Recall. Whenever False Negatives and False Positives are crucial, this technique is used. We need to avoid false negatives so we don't miss out on people likely to develop diabetes and delay their illness; we also want to avoid false positives so we don't increase unnecessary worry. Therefore f1_weighted is our metric to evaluate models.

### 11.1.2 Modelling

### Decision Trees

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
grid_values = {'criterion':['gini','entropy'],'splitter':['best','random'],'max_depth':[1,10,15,20,25,50],'random_state':[4]}
dtree = DecisionTreeClassifier()
grid_search = GridSearchCV(dtree,grid_values,cv = 5, scoring='f1_weighted')
grid_result = grid_search.fit(X_train,y_train)
grid_result.best_params_
```

```python
dtree = DecisionTreeClassifier(criterion = 'entropy',max_depth = 25,splitter = 'random',random_state = 4)
dtree.fit(X_train,y_train)
# from sklearn.tree import DecisionTreeClassifier,export_graphviz
# import pydotplus
# with open('c.png','wb') as fw:    
#     fw.write(pydotplus.graph_from_dot_data(export_graphviz(dtree)).create_png())
# from IPython.display import Image
Image('c.png')
```

```python
predictions = dtree.predict(X_train)
print('F1_Score_train=',format(metrics.f1_score(y_train, predictions,average='weighted')))
```

```python
predictions = dtree.predict(X_test)
print('F1_Score_test=',format(metrics.f1_score(y_test, predictions,average='weighted')))
```

```python
print(classification_report(y_test, predictions))
```
